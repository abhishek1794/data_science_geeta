```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
#loading datasets
df_train = pd.read_csv('train.csv')
df_test  = pd.read_csv('test.csv')
#concatenation
df_combined = pd.concat([df_train, df_test])
```

```python
#data mapping
df_combined['Gender'] = df_combined.Gender.map({'Male':0,'Female':1})
```

```python
#date to datetime
df_combined['DOB']= pd.to_datetime(df_combined['DOB'])
df_combined['DOB'].dtypes
#now we can do operations with date
df_combined['age'] = 2020 - df_combined.DOB.dt.year
df_combined['age']=np.where(df_combined['age'] <0, '0', df_combined['age']) #if <0.. then '0' otherwise age
df_combined.head()
```
